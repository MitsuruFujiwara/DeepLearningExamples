```python
import pandas as pd
import numpy as np
from sklearn.externals import joblib
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score, roc_curve, auc, f1_score, precision_score, recall_score
from sklearn.svm import SVC
from keras.models import load_model
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

Load Validation Dataset

```python
df1 = pd.read_hdf('data.h5', key='data1_test', mode='r')
df2 = pd.read_hdf('data.h5', key='data2_test', mode='r')
df3 = pd.read_hdf('data.h5', key='data3_test', mode='r')
df4 = pd.read_hdf('data.h5', key='data4_test', mode='r')
df5 = pd.read_hdf('data.h5', key='data5_test', mode='r')

X1, Y1 = np.array(df1.drop('class', axis=1)), np.array(df1['class'])
X2, Y2 = np.array(df2.drop('class', axis=1)), np.array(df2['class'])
X3, Y3 = np.array(df3.drop('class', axis=1)), np.array(df3['class'])
X4, Y4 = np.array(df4.drop('class', axis=1)), np.array(df4['class'])
X5, Y5 = np.array(df5.drop('class', axis=1)), np.array(df5['class'])
```

Load Training Dataset

```python
df1_t = pd.read_hdf('data.h5', key='data1_train', mode='r')
df2_t = pd.read_hdf('data.h5', key='data2_train', mode='r')
df3_t = pd.read_hdf('data.h5', key='data3_train', mode='r')
df4_t = pd.read_hdf('data.h5', key='data4_train', mode='r')
df5_t = pd.read_hdf('data.h5', key='data5_train', mode='r')

X1_t, Y1_t = np.array(df1_t.drop('class', axis=1)), np.array(df1_t['class'])
X2_t, Y2_t = np.array(df2_t.drop('class', axis=1)), np.array(df2_t['class'])
X3_t, Y3_t = np.array(df3_t.drop('class', axis=1)), np.array(df3_t['class'])
X4_t, Y4_t = np.array(df4_t.drop('class', axis=1)), np.array(df4_t['class'])
X5_t, Y5_t = np.array(df5_t.drop('class', axis=1)), np.array(df5_t['class'])
```

Load Model

```python
lr1 = joblib.load('lr1.pkl') 
lr2 = joblib.load('lr2.pkl') 
lr3 = joblib.load('lr3.pkl') 
lr4 = joblib.load('lr4.pkl') 
lr5 = joblib.load('lr5.pkl') 

mlp1 = joblib.load('mlp1.pkl') 
mlp2 = joblib.load('mlp2.pkl') 
mlp3 = joblib.load('mlp3.pkl') 
mlp4 = joblib.load('mlp4.pkl') 
mlp5 = joblib.load('mlp5.pkl') 

dt1 = joblib.load('dt1.pkl') 
dt2 = joblib.load('dt2.pkl') 
dt3 = joblib.load('dt3.pkl') 
dt4 = joblib.load('dt4.pkl') 
dt5 = joblib.load('dt5.pkl') 

svm1 = joblib.load('svm1.pkl')
svm2 = joblib.load('svm2.pkl')
svm3 = joblib.load('svm3.pkl')
svm4 = joblib.load('svm4.pkl')
svm5 = joblib.load('svm5.pkl')

xgb1 = joblib.load('xgb1.pkl')
xgb2 = joblib.load('xgb2.pkl')
xgb3 = joblib.load('xgb3.pkl')
xgb4 = joblib.load('xgb4.pkl')
xgb5 = joblib.load('xgb5.pkl')

rf1 = joblib.load('rf1.pkl')
rf2 = joblib.load('rf2.pkl')
rf3 = joblib.load('rf3.pkl')
rf4 = joblib.load('rf4.pkl')
rf5 = joblib.load('rf5.pkl')

dnn1_ae3=load_model('DNN1_ae3.h5')
dnn2_ae3=load_model('DNN2_ae3.h5')
dnn3_ae3=load_model('DNN3_ae3.h5')
dnn4_ae3=load_model('DNN4_ae3.h5')
dnn5_ae3=load_model('DNN5_ae3.h5')

dnn1_ae4=load_model('DNN1_ae4.h5')
dnn2_ae4=load_model('DNN2_ae4.h5')
dnn3_ae4=load_model('DNN3_ae4.h5')
dnn4_ae4=load_model('DNN4_ae4.h5')
dnn5_ae4=load_model('DNN5_ae4.h5')

dnn1_ae5=load_model('DNN1_ae5.h5')
dnn2_ae5=load_model('DNN2_ae5.h5')
dnn3_ae5=load_model('DNN3_ae5.h5')
dnn4_ae5=load_model('DNN4_ae5.h5')
dnn5_ae5=load_model('DNN5_ae5.h5')
```

Compile DNN

```python
model1=[lr1, mlp1, dt1, svm1, xgb1, rf1]
model2=[lr2, mlp2, dt2, svm2, xgb2, rf2]
model3=[lr3, mlp3, dt3, svm3, xgb3, rf3]
model4=[lr4, mlp4, dt4, svm4, xgb4, rf4]
model5=[lr5, mlp5, dt5, svm5, xgb5, rf5]
model = [model1, model2, model3, model4, model5]

model_dnn1=[dnn1_ae3, dnn1_ae4, dnn1_ae5]
model_dnn2=[dnn2_ae3, dnn2_ae4, dnn2_ae5]
model_dnn3=[dnn3_ae3, dnn3_ae4, dnn3_ae5]
model_dnn4=[dnn4_ae3, dnn4_ae4, dnn4_ae5]
model_dnn5=[dnn5_ae3, dnn5_ae4, dnn5_ae5]
model_dnn = [model_dnn1, model_dnn2, model_dnn3, model_dnn4, model_dnn5]

model_name= ['lr', 'mlp', 'dt', 'svm', 'xgb', 'rf', 'dnn3', 'dnn4', 'dnn5']
model_name_dnn = ['dnn3', 'dnn4', 'dnn5']

result1={}
result2={}
result3={}
result4={}
result5={}

for s, t in zip(model1, model_name):
    res_model = {}
    res_model['AUC']=roc_auc_score(Y1, s.predict_proba(X1)[:,1])
    res_model['Train AUC']=roc_auc_score(Y1_t, s.predict_proba(X1_t)[:,1])
    result1[t] = res_model
    
for s, t in zip(model2, model_name):
    res_model = {}
    res_model['AUC']=roc_auc_score(Y2, s.predict_proba(X2)[:,1])
    res_model['Train AUC']=roc_auc_score(Y2_t, s.predict_proba(X2_t)[:,1])
    result2[t] = res_model

for s, t in zip(model3, model_name):
    res_model = {}
    res_model['AUC']=roc_auc_score(Y3, s.predict_proba(X3)[:,1])
    res_model['Train AUC']=roc_auc_score(Y3_t, s.predict_proba(X3_t)[:,1])
    result3[t] = res_model

for s, t in zip(model4, model_name):
    res_model = {}
    res_model['AUC']=roc_auc_score(Y4, s.predict_proba(X4)[:,1])
    res_model['Train AUC']=roc_auc_score(Y4_t, s.predict_proba(X4_t)[:,1])
    result4[t] = res_model

for s, t in zip(model5, model_name):
    res_model = {}
    res_model['AUC']=roc_auc_score(Y5, s.predict_proba(X5)[:,1])
    res_model['Train AUC']=roc_auc_score(Y5_t, s.predict_proba(X5_t)[:,1])
    result5[t] = res_model
```
